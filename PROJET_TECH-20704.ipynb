```python
def title_of_type(df, _type_):
    columns = [df.columns[i] for i, x in enumerate([df[i].dtypes==_type_ for i in df.columns]) if x == True]
    return columns
```

```python
print('float : {0}'.format(title_of_type(df, float)))
print('int : {0}'.format(title_of_type(df, int)))
print('str : {0}'.format(title_of_type(df, str)))

# print('str : {0}'.format(title_of_type(df, datetime.datetime)
# print(type(pd.Timestamp(ts_input=df['DATE'].iloc[0])))
```
